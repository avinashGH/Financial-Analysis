```python
ar = pd.read_csv('Downloads/ARE&M.BO.csv',index_col=0)
```

**Select single column from a data frame**

```python
print(ar['Close'])
```

```python
## Select Multiple Columns from DF
ar[['Open','Close']]
```

## Create a new column in a DF


```python
## shift is used to move a whole column
ar['Price']= ar['Close'].shift(-1)
```

```python
ar
```

### Create Price Diff

```python
## Tomorrow - Today
ar['PriceDiff'] = ar['Price']-ar['Close']
```

```python
ar
```

### Create a daily return Column


```python
## (Tmorrow price - today price)/today price = pricediff/close today
ar['D Return'] = ar['PriceDiff']/ar['Close']
```

```python
ar.drop('D Reture',axis=1,inplace =True)
```

```python
ar.head()
```

### Create Direction Column

### Rules - PriceDiff > 0 --> up 1
### PriceDiff < 0 --> down -1

## List comprehension

```python
ar.index
```

```python
ar['Direction'] = [1 if ar.loc[ei,'PriceDiff'] > 0  else -1 for ei in ar.index]
```

```python
ar.tail()
```

## Create a Moving Averager Column

```python
## For 3 days - Today,Yesterday & Day Before Yesterday
ar['Average3'] = (ar['Close'] + ar['Close'].shift(1) + ar['Close'].shift(2))/3  
## shift(n) -> shift whole column downward by n row
```

```python
ar.tail()
```

```python
### missing values in df is as NaN 
```

## Calculate moving average using `.rolling()`

```python
## rolling is a function to provide us window calculation
ar['MA40'] = ar['Close'].rolling(40).mean()
ar['MA100'] = ar['Close'].rolling(100).mean()
```

```python
ar.head()
```

## Plot Moving Average

```python
plt.figure(figsize=(8,4))
ar['Close'].plot()
ar['MA40'].plot()
ar['MA100'].plot()
plt.show()
```
